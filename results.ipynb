```python
T_vals = [1, 10, 50]
bias    = [1.5847, 0.4340, 0.1758]
rmse    = [1.7684, 0.9291, 0.7937]
kl      = [0.637,  0.035,  0.106]
cal     = [0.083,  0.323,  0.538]
std     = [0.785,  0.822,  0.774]

fig, axes = plt.subplots(1, 3, figsize=(13, 4))

axes[0].plot(T_vals, bias, 'o-', color='steelblue', label='|Bias|')
axes[0].plot(T_vals, rmse, 's--', color='coral',    label='RMSE')
axes[0].set_xlabel('T'); axes[0].set_ylabel('Value')
axes[0].set_title('Bias & RMSE vs. T')
axes[0].legend(); axes[0].grid(True, alpha=0.3)
axes[0].set_xticks(T_vals)

axes[1].plot(T_vals, kl, 'o-', color='steelblue')
axes[1].set_xlabel('T'); axes[1].set_ylabel('KL divergence')
axes[1].set_title('KL(posterior || prior) vs. T')
axes[1].grid(True, alpha=0.3); axes[1].set_xticks(T_vals)

axes[2].plot(T_vals, cal, 'o-', color='steelblue')
axes[2].set_xlabel('T'); axes[2].set_ylabel('Calibration Score')
axes[2].set_title('Calibration Score vs. T')
axes[2].set_ylim(0, 0.7); axes[2].grid(True, alpha=0.3)
axes[2].set_xticks(T_vals)

fig.suptitle('GaussianRandomWalk — Inference Quality vs. Observation Length',
             fontsize=12)
plt.tight_layout()
plt.savefig('GRW_summary_metrics.png', dpi=150, bbox_inches='tight')
plt.show()
```
